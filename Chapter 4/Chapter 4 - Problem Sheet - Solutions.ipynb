## Solution 1

### Case study: Bayesian M&M candies

We need Bayesian inference to answer this problem.

Let us assume that the yellow one comes from bag 1 and the green one from bag 2.

There are two possible outcomes, $H = \{A, B\}$:

- $A$: Bag 1 (yellow) is vintage and bag 2 (green) is new
- $B$: Bag 1 (yellow) is new and bag 2 (green) is vintage

Bayes theorem says:

\begin{equation}
  P(H \mid D) = \frac{P(H) P(D \mid H)}{P(D)}
\end{equation}

Let us construct a table in which we fill in the terms of Bayes theorem:

| H    | Prior | Likelihood   | P(H) P(D;H) | Posterior     |
| :--: | :---: | :----------: | :---------: | :-----------: |
| A    | 0.5   | (0.20)(0.20) | 0.020       | 0.020 / 0.027 |
| B    | 0.5   | (0.14)(0.10) | 0.007       | 0.007 / 0.027 |

Therefore, $P(A) = 74.1\%$ and $P(B) = 25.9\%$.

I would definitely choose to eat from bag 2, which is the one from which the green M&M came.

_This was an example of MAP for discrete data._

## Solution 2

### MLE of a Gaussian with unknown mean $\mu$ and standard deviation $\sigma$

#### Task 1
We can express the likelihood of a single event as:
\begin{equation}
  \mathcal{L}(x_i \mid \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} \exp{\left( \frac{-(x_i-\mu)^2}{2\sigma^2} \right)} \,.
\end{equation}

#### Task 2
For the full dataset, the likelihood becomes:
\begin{equation}
  \mathcal{L}(x \mid \mu, \sigma) = \prod_i \mathcal{L}(x_i \mid \mu, \sigma) \,.
\end{equation}

We therefore obtain the joint log-likelihood of the dataset as:
\begin{eqnarray}
  \ln \mathcal{L} &=& \sum \ln \mathcal{L}(x_i | \mu, \sigma) \\
                  &=& -N \ln \sigma - \frac{N}{2} \ln (2\pi) -\frac{1}{2 \sigma^2} \sum \left( x_i - \mu \right)^2 \,.
\end{eqnarray}

The second term from the equation above is a constant which only depends on the data.<br><br>

#### Task 3
Since neither $\mu$ nor $\sigma$ are known, we need to simultaneously find the MLE $\widehat{\mu}$ and $\widehat{\sigma}$:

\begin{eqnarray}
  \left[ \frac{\partial \ln \mathcal{L}}{\partial \mu} \right]_{\widehat{\mu},\widehat{\sigma}} &=& \frac{1}{\sigma^2} \sum \left( x_i - \widehat{\mu} \right) = 0 \,, \\
  \left[ \frac{\partial \ln \mathcal{L}}{\partial \sigma} \right]_{\widehat{\mu},\widehat{\sigma}} &=& -\frac{N}{\widehat{\sigma}} + \frac{1}{\widehat{\sigma}^3} \sum \left( x_i - \widehat{\mu} \right)^2 = 0 \,.
\end{eqnarray}

From the first equation we find:
\begin{equation}
  \widehat{\mu} = \frac{1}{N} \sum x_i \equiv \langle x \rangle \,,
\end{equation}

which we recognise as the sample mean.

From the second equation, after substituting $\langle x \rangle = \widehat{\mu}$ we obtain:
\begin{equation}
  \widehat{\sigma}^2 = \frac{1}{N} \sum \left( x_i - \langle x \rangle \right)^2 \, \equiv \sigma_{\rm uncorrected}^2 \,.
\end{equation}

This last result tells us that the MLE $\widehat{\sigma}$ is the well-known uncorrected standard deviation of a sample set, which we know is a biased estimator. However, in the limit $N \to \infty$, the difference between $N$ and $N-1$ becomes negligible.

#### Task 4
For MLE problems with multiple parameters, we can calculate the variance on the estimated parameters using the Fisher information matrix
\begin{equation}
  \mathbf{V}(\widehat{a}) = \left(-\mathbf{H}(\widehat{a})\right)^{-1} \,.
\end{equation}

Let us work out the components of $\mathbf{H}(\widehat{a})$
\begin{eqnarray}
    H_{\widehat{\mu},\widehat{\mu}} &=& \left. \frac{\partial^2 \ln \mathcal{L}}{\partial \mu^2} \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \left. \frac{\partial}{\partial \mu} \left( \frac{1}{\sigma^2} \sum (x_i-\mu) \right) \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& -\frac{N}{\widehat{\sigma}^2} \,,
\end{eqnarray}

\begin{eqnarray}
    H_{\widehat{\sigma},\widehat{\sigma}} &=& \left. \frac{\partial^2 \ln \mathcal{L}}{\partial \sigma^2} \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \left. \frac{\partial}{\partial \sigma} \left( \frac{1}{\sigma^3} \sum (x_i-\mu)^2 - \frac{N}{\sigma} \right) \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \left. \frac{-3}{\sigma^4} \sum (x_i-\mu)^2 + \frac{N}{\sigma^2} \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \frac{-2N}{\widehat{\sigma}^2} \,,
\end{eqnarray}

\begin{eqnarray}
    H_{\widehat{\mu},\widehat{\sigma}} &=& \left. \frac{\partial^2 \ln \mathcal{L}}{\partial \mu \partial \sigma} \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \left. \frac{\partial}{\partial \sigma} \left( \frac{1}{\sigma^2} \sum (x_i-\mu) \right) \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \left. \frac{-2}{\sigma^3} \sum (x_i-\mu) \right\rvert_{\widehat{\mu},\widehat{\sigma}} \\
                     &=& \frac{2}{\widehat{\sigma}^3} \left( \left( \sum x_i \right) - N\widehat{\mu} \right) \\
                     &=& 0 \,.
\end{eqnarray}

Hence
\begin{eqnarray}
    \mathbf{H} &=&  
        \begin{bmatrix}
        \frac{-N}{\widehat{\sigma}^2} & 0 \\
        0 & \frac{-2N}{\widehat{\sigma}^2} \\
        \end{bmatrix} \,.
\end{eqnarray}

Finally
\begin{eqnarray}
    \mathbf{V}(\widehat{a}) &=& \left(-\mathbf{H}(\widehat{a})\right)^{-1} \\
        &=&
        \begin{bmatrix}
        \frac{\widehat{\sigma}^2}{N} & 0 \\
        0 & \frac{\widehat{\sigma}^2}{2N} \\
        \end{bmatrix} \,.
\end{eqnarray}
