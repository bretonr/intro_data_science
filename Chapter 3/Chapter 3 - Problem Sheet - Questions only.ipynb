### Variance with unknown mean

Let us now consider the variance for the case where the true mean $\mu$ is unknown and instead replaced by the estimate $\widehat{\mu} = \langle x \rangle$, which is nothing else than the sample mean:
\begin{equation}
  \widehat{V(x)} = \frac{1}{N} \sum \left(X_i - \langle X \rangle\right)^2 = \frac{1}{N} \sum \left((X_i^2 - \langle X \rangle^2\right) \,.
\end{equation}

Show that this is a biased estimator.
